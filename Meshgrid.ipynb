```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook

# The point of meshgrid is to make a grid of points, to which we can then make a contour plot by taking 
# values at many 2D (x,y) points. 

# Without meshgrid: make two seperate empty arrays, and fill them out using a for loop

x = np.zeros([5,5])
y = np.zeros([5,5])
for i in range(5):
    for j in range(5):
        x[i,j] = i
        y[i,j] = j

plt.figure()
plt.plot(x,y, marker='.', color='k', linestyle='none')
plt.show()
```

```python
# now, doing the same using meshgrid:
xvalues = np.arange(1,6,1)
yvalues = np.arange(1,6,1)

xx,yy = np.meshgrid(xvalues, yvalues)
plt.figure()
plt.plot(xx,yy, marker ='.', color='k', linestyle='none')
plt.show()

print(type(xx))
# It's much shorter to type, and much faster; we should avoid for loops when we can
```

```python
# let's try using our grid to visualize what the function f(x,y) = sin(x**2 + y**2) / (x**2 + y**2) looks like. 
# To do this without making a 3D plot, we can use plt.imshow() and input our grid that we made.

xx,yy = np.meshgrid(xvalues, yvalues)
# Note: this function is undefined at (0,0). Be careful about these things
def func(x,y):
    return np.sin(x**2 + y**2)/(x**2 + y**2)

# origin='lower' is required to put the origin in the bottom left corner. By default, it's in the upper left corner.
plt.figure()
plt.imshow(func(xx,yy),origin='lower')
plt.colorbar()
plt.show()

# cool
```

```python
# lets try this using more points:
xs, ys = np.meshgrid(np.arange(1,6,0.01),np.arange(1,6,0.01))
plt.figure()
plt.imshow(func(xs,ys),origin='lower')
plt.colorbar()
plt.show()
```
